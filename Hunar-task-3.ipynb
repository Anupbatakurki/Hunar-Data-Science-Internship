```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'breast cancer.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Encode the target variable
data['diagnosis'] = LabelEncoder().fit_transform(data['diagnosis'])

# Separate features and target variable
X = data.drop(columns=['id', 'diagnosis'])
y = data['diagnosis']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

The data has been successfully split into training and testing sets. The training set consists of 455 samples, and the testing set consists of 114 samples.
Next, let's start with k=5, a common default value for the number of neighbors, and train the K-Nearest Neighbors (KNN) model. We will then evaluate the model's performance using the testing set.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Initialize the KNN model with k=5
k = 5
knn = KNeighborsClassifier(n_neighbors=k)

# Train the model on the training data
knn.fit(X_train, y_train)

# Predict on the testing data
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

accuracy, precision, recall, f1

```

The initial evaluation of the KNN model with \( k = 5 \) yields the following performance metrics:

- Accuracy = 94.74%
- Precision = 93.02%
- Recall = 93.02%
- F1 Score = 93.02%

These results are quite promising.
Next




```python
import matplotlib.pyplot as plt

# Evaluate model performance for different values of k
k_values = range(1, 20)
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))

# Plot the performance metrics for different values of k
plt.figure(figsize=(12, 6))
plt.plot(k_values, accuracy_scores, label='Accuracy', marker='o')
plt.plot(k_values, precision_scores, label='Precision', marker='o')
plt.plot(k_values, recall_scores, label='Recall', marker='o')
plt.plot(k_values, f1_scores, label='F1 Score', marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Score')
plt.title('KNN Performance Metrics for Different k Values')
plt.legend()
plt.grid(True)
plt.show()
```
